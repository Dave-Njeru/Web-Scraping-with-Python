```python
# Import needed libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
from google.colab import files
```

```python
# Fetch website data
url = 'https://www.scrapethissite.com/pages/forms/'
webpage = requests.get(url)
```

```python
# Use BeatifulSoup
soup = BeautifulSoup(webpage.text, 'html')
print(soup)
```

```python
# Extract the table with the Hockey Scores
hockey_table = soup.find('table', class_='table')
# print(hockey_table)
```

```python
# Extract the column headings
table_titles = hockey_table.find_all('th')
hockey_table_titles = [title.text.strip() for title in table_titles]

# Alternative code for the list comprehesion
# hockey_table_titles = []
# for title in table_titles:
#     hockey_table_titles.append(title.text.strip())

# print(hockey_table_titles)
```

```python
# Save the column headings onto a Pandas DataFrame
hockey_data_frame = pd.DataFrame(columns=hockey_table_titles)
# print(hockey_data_frame)
```

```python
# Extract data row by row
# 1. Get all rows
hockey_table_rows = hockey_table.find_all('tr')

# 2. Loop through each row while stripping and data into the dataframe
for row in hockey_table_rows[1:]: # Start from the second row to skip headers
    hockey_table_data = row.find_all('td')
    hockey_table_row_data = [data.text.strip() for data in hockey_table_data]
    # Append the row data to the DataFrame
    # Using pd.concat for efficient appending in a loop
    length = len(hockey_data_frame)
    hockey_data_frame.loc[length] = hockey_table_row_data

# display(hockey_data_frame)
```

```python
# Save results to a .csv file
hockey_data_frame.to_csv('./hockey_scores.csv', index=False)
# Download the .csv file to local machine
files.download('./hockey_scores.csv')
```
